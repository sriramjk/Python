```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, '*')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
```

```python
sns.set()
```

```python
# Create a pandas DataFrame with the data
data = pd.DataFrame({'x': x, 'y': y})

# Use the 'x' and 'y' columns as arguments for lineplot
sns.lineplot(x='x', y='y', data=data)
plt.show()
```

```python
data = pd.read_csv('mobile_cleaned.csv')
data.head()
```

```python
data.shape
```

```python
ax = sns.scatterplot(x='stand_by_time', y='battery_capacity', data=data)
plt.show()
```

```python
ax = sns.scatterplot(x='stand_by_time', y='battery_capacity', data=data, hue='thickness')
plt.show()
```

```python
data.columns
```

```python
data['thickness'].unique()
```

```python
ax = sns.distplot(data['stand_by_time'])
plt.show()
```

```python
ax = sns.distplot(data['stand_by_time'], kde=False)
plt.show()
```

```python
ax = sns.boxplot(x='is_liked', y='battery_capacity', data=data)
```
